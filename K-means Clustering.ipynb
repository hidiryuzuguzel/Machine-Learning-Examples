## Load data

```matlab
clear all; close all;
load MFCCfeatures
input_feature = VowelFeatures(:,[4 5 8]);
```

```matlab
[means,inds] = MyKmeans(input_feature,8,'Euclidean'); 
```

## matlab built-in

```matlab
[idx,ctrs] = kmeans(input_feature,8,'Distance','sqEuclidean');
color = {'y.','m.','c.','r.','g.','b.','k.','rd'};
figure,
hold on
for k=1:8
    plot3(input_feature(idx==k,1),input_feature(idx==k,2),input_feature(idx==k,3),color{k})
    plot3(ctrs(:,1),ctrs(:,2),ctrs(:,3),'kx','Markersize',12); 
   plot3(ctrs(:,1),ctrs(:,2),ctrs(:,3),'ko','Markersize',12); 
end
hold off
```
